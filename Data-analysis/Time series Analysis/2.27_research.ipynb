```python
from statsmodels.tsa.seasonal import seasonal_decompose

series = data_set_1[100000:120000:,1]
result = seasonal_decompose(series, model='additive', freq=1)
print(result.trend)
print(result.seasonal)
print(result.resid)
print(result.observed)
result.plot()
plt.show()
```
