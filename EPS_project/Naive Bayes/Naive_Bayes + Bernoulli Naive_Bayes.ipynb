```python
from sklearn import preprocessing
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Total_datas = pd.read_csv('C:/Users/hml76/OneDrive/바탕 화면/Capacitive_sensor_research/실험데이터/Total_datas.csv')
Total_labels =  pd.read_csv('C:/Users/hml76/OneDrive/바탕 화면/Capacitive_sensor_research/실험데이터/Total_labels.csv')
```

```python
X_train, X_test, y_train, y_test = train_test_split(Total_datas, Total_labels, test_size=0.3, random_state=50)

model = GaussianNB()

# Train the model using the training sets
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print("훈련 세트 정확도: {:.3f}".format(model.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(model.score(X_test, y_test)))
print("테스트 세트 정확도: {:.3f}".format(model.score(y_test,y_pred)))
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
