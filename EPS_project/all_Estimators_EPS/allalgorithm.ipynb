```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame as df
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import warnings
from sklearn.utils.testing import all_estimators
```

```python
Total_datas = pd.read_csv('C:/Users/hml76/OneDrive/바탕 화면/Capacitive_sensor_research/실험데이터/Total_datas.csv')
Total_labels =  pd.read_csv('C:/Users/hml76/OneDrive/바탕 화면/Capacitive_sensor_research/실험데이터/Total_labels.csv')
```

```python
X_train, X_test, y_train, y_test = train_test_split(Total_datas, Total_labels, test_size=0.3, shuffle=True)
```

```python
#warnings.filterwarnings('ignore')
allAlgorithms = all_estimators(type_filter="classifier")
```

```python
for name, algorithm in allAlgorithms:
    if hasattr(algorithm, 'predict_proba'):
        clf = algorithm()
        clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
print(name, "의 정확도 : ", accuracy_score(y_test,y_pred))
```
