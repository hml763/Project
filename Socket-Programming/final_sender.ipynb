```python
import socket 

Sender_IP = '127.0.0.1'
Receiver_IP = '127.0.0.1'
Sender_Router_port = 9998
Receiver_Router_port = 9999
Data = 'Data'
seq_num = 0
message_num = 150  #message content : 1~150, 250, 450, 650, 850, 1000

sender_socket = socket.socket(socket.AF_INET,socket.SOCK_STREAM) 
sender_socket.connect((Sender_IP, Sender_Router_port)) 


def start_sending_packets(seq_num):
    for i in range(1,message_num+1):  
        seq_num = seq_num + 1
        packet = [Sender_IP,Sender_Router_port,Receiver_IP,Receiver_Router_port,seq_num,Data]  

        sender_socket.send( repr(packet).encode()) 
        data = sender_socket.recv(1024) 

        print('Packet sent to the receiver :',repr(data.decode("utf-8"))) 
    
    recur(seq_num)

    
def recur(seq_num):
    print('\nPlease enter "start" to start the program, or "exit" to end program')
    IN=input()
    
    if(IN == 'exit'):
        sender_socket.send('exit'.encode('utf-8')) 
        print("\nThank you! :)\n\n")
        sender_socket.close()

    elif(IN == 'start'):
        seq_num = 0
        start_sending_packets(seq_num)

    else:
        print('\nYou entered the wrong letter, please try again\n')
        recur(seq_num)

print("Assignment 1, programmed by Hunmin Lee\n\n\n")
recur(seq_num)    
```
