```python
import socket 
from _thread import *
import time
import sys

buffer = []
recccc = []


def threaded(sender_socket, addr): 
    print('Connected by :', addr[0], ':', addr[1]) 

    
    while True: 
        try:
            data = sender_socket.recv(1024) 
            if(data.decode() == 'exit'):
                receiver_socket.send('exit'.encode('utf-8'))
                router_socket.close()
                            
            if not data: 
                print('Disconnected by ' + addr[0],':',addr[1])
                print('router buffer : ', buffer) 
                break

            buffer.append(data.decode())
            sender_socket.send(data) 
            time.sleep(0.001)
            print('packet saving to buffer, and forward to receiver :', data.decode())
            time.sleep(0.001)
            receiver_socket.send(data)
            
            received_receiver = receiver_socket.recv(1024)
            
            if(received_receiver.decode('utf-8') == 'NACK'):
                print("NACK!")
                time.sleep(0.001)
                receiver_socket.send(data)
                time.sleep(0.001)
            
            
        except ConnectionResetError as e:
            print('Error_Disconnected by ' + addr[0],':',) 
            break
             
    sender_socket.close() 

    
HOST = '127.0.0.1'
PORT = 9998
PORT2 = 9999

router_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM) 
router_socket.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
router_socket.bind((HOST, PORT)) 
router_socket.listen() 


receiver_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM) 
receiver_socket.connect((HOST, PORT2)) 

print('server start')


while True: 
    print('waiting for packet to arrive from sender ... ') 
    sender_socket, addr = router_socket.accept()  
    start_new_thread(threaded, (sender_socket, addr)) 
    
receiver_socket.close()
```
